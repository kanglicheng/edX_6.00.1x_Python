```python
# Problem 1-4: For large values of n, an algorithm that takes 20000n^2 
# steps has better time complexity (takes less time) than one that takes 0.001n^5 steps (T/F)

%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

a = 20000
b = 0.001

n = np.linspace(0.01, 10**10, 100)

aVals = a * n**2
bVals = b * n**5

plt.plot(n, aVals, 'r--', label='a')
plt.plot(n, bVals, 'b--', label='b')
plt.legend()
```
