```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import log
```

```python
xVals = []
const = []
lg = []
linear = []
loglin = []
polyQuad = []
exp = []

for x in range(1, 7):
    xVals.append(x)
    const.append(1)
    lg.append(log(x))
    linear.append(x)
    loglin.append(x * log(x))
    polyQuad.append(x**2)
    exp.append(2**x)
```

```python
plt.figure('complexities', figsize=(15,10))
plt.clf()

plt.plot(xVals, const, 'r', label='Constant')
plt.plot(xVals, lg, 'b', label='Logarithmic')
plt.plot(xVals, linear, 'k', label='Linear')
plt.plot(xVals, loglin, 'g', label='Log-Linear')
plt.plot(xVals, polyQuad, 'purple', label='Quadratic')
plt.plot(xVals, exp, 'orange', label='Exponential')

plt.legend(loc=0, fontsize=14)
plt.title('Algorithmic Complexity Comparison', fontsize=18)
plt.xlabel('Input Magnitude N', fontsize=14)
plt.ylabel('Number of Computations', fontsize=14)
```
